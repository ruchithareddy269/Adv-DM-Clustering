```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data points
data_points=np.array([
    [3, 7],
    [2, 6],
    [2, 2],
    [5, 8],
    [5, 5],
    [6, 6],
    [8, 4],
    [5, 2],
    [7, 3],
    [12, 8],
    [10, 6]
])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

#Single-Link
single_linkage_matrix=linkage(data_points,method='single')

#Average-Link
average_linkage_matrix=linkage(data_points,method='average')

#Complete-Link
complete_linkage_matrix=linkage(data_points,method='complete')

#visualizing the results using dendrograms
fig,ax=plt.subplots(3,1,figsize=(9,10))

#Single-Link Dendrogram

dendrogram(single_linkage_matrix, ax=ax[0], labels=[str(point) for point in data_points])
ax[0].set_title('Single Link Hierarchical Clustering')
ax[0].set_xlabel('Data Points')
ax[0].set_ylabel('Distance')

#Average-Link Dendrogram
dendrogram(average_linkage_matrix, ax=ax[1], labels=[str(point) for point in data_points])
ax[1].set_title('Average Link Hierarchical Clustering')
ax[1].set_xlabel('Data Points')
ax[1].set_ylabel('Distance')

#Complete-Link Dendrogram
dendrogram(complete_linkage_matrix, ax=ax[2], labels=[str(point) for point in data_points])
ax[2].set_title('Complete Link Hierarchical Clustering')
ax[2].set_xlabel('Data Points')
ax[2].set_ylabel('Distance')

plt.tight_layout()
plt.show()
```
